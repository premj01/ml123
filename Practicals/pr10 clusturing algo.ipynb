```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans, AgglomerativeClustering
```

```python
iris = load_iris()
X = pd.DataFrame(iris.data, columns=iris.feature_names)
```

```python
X_2d = X[['sepal length (cm)', 'petal length (cm)']]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_2d)

kmeans = KMeans(n_clusters=3)
kmeans_labels = kmeans.fit_predict(X_scaled)
```

```python
# Step 4. Apply Hierarchical Clustering
hier = AgglomerativeClustering(n_clusters=3)
hier_labels = hier.fit_predict(X_scaled)
```

```python
# Step 5. Plot Results (Scatter Plots)
plt.figure(figsize=(10,4))

# K-Means
plt.subplot(1,2,1)
plt.scatter(X_scaled[:,0], X_scaled[:,1], c=kmeans_labels, cmap='viridis', s=50)
plt.title("K-Means Clustering")
plt.xlabel("Sepal Length (scaled)")
plt.ylabel("Petal Length (scaled)")

# Hierarchical
plt.subplot(1,2,2)
plt.scatter(X_scaled[:,0], X_scaled[:,1], c=hier_labels, cmap='plasma', s=50)
plt.title("Hierarchical Clustering")
plt.xlabel("Sepal Length (scaled)")
plt.ylabel("Petal Length (scaled)")

plt.tight_layout()
plt.show()
```
